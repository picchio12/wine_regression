```python
def boxplot_n_histogram(dataset, column):
    fig, axes = plt.subplots(nrows=1, ncols=2, sharex=False, sharey=False, figsize=(15, 7))
    axes[0].boxplot(dataset[column], showmeans=True)
    axes[1].hist(dataset[column], 
                 # bins=len(dataset[column].unique())
    )
    
    axes[0].set_title(column)
    axes[1].set_title(column)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import KFold, cross_validate, GridSearchCV, train_test_split

x = data.drop(columns="quality")
y = data["quality"]
kf = KFold(5, shuffle=True, random_state=34)
```

## Regressione lineare con regolarizzazione

```python
linear_model = Pipeline([
    ("scaler", StandardScaler()), 
    ("linreg", LinearRegression()
     )])

cv_res = cross_validate(linear_model, x, y, cv=kf, return_train_score=True)
cv_frame = pandas.DataFrame(cv_res)
```

```python
cv_frame
```

## Ridge con grid search

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=34)

ridge_model = Pipeline([
    ("scaler", StandardScaler()),
    ("ridge", Ridge())
])
grid = {"ridge__alpha":[0.1, 1, 10, 30, 50]}
gs_res = GridSearchCV(ridge_model, param_grid=grid, cv=5)
gs_res.fit(x_train, y_train)
```

```python
pandas.DataFrame(gs_res.cv_results_)
```

Al crescere di alpha non si ottengono miglioramenti significativi rispetto alla semplice regressione lineare.
Notiamo anzi un piccolo abbassamento della precisione utilizzando il metodo Ridge.

La precisione dei primi due modelli testati non  soddisfacente.
Tentiamo di con una regressione Lasso, siccome notiamo un alto indice di correlazione tra alcune della variabili indipendenti.

## Regressione Lasso con grid search

```python
ridge_model = Pipeline([
    ("scaler", StandardScaler()),
    ("lasso", Lasso())
])
grid = {"lasso__alpha":[0.1, 0.4, 0.7, 1]}
gs_res = GridSearchCV(ridge_model, param_grid=grid, cv=5)
gs_res.fit(x_train, y_train)
```

```python
pandas.DataFrame(gs_res.cv_results_)
```
